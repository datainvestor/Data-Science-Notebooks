```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
import time
headers = {'User-Agent': 'Mozilla/5.0'}
```

```python
url="https://aplikacje.nfz.gov.pl/umowy/Provider/Index?ROK=2018&OW=01&ServiceType=00&OrthopedicSupply=False&page=1"
page=requests.get(url, headers=headers)
soup=BeautifulSoup(page.content, 'html.parser')
content=soup.findAll("tr")
```

```python
list1=[] #parsed rows
mainl=[] #main list
al=[] #list for first column
for r in content:
    cols=r.find_all('td')
    atag=r.find('a')
    al.append(atag.get_text().strip().split(' ', 1)[0])
    for td in cols:       
        try:
            text = td.find(text=True)      
            list1.append(text.strip())
        except:
            continue
            
#remove first item from al list
al=al[1:]
al = list(map(lambda s: s.strip(), al)) #clean strings
al

#remove nones from list
while '' in list1:
    list1.remove('')
    
#prepare for dataframe
for i in range(7):
    mainl.append(list1[i::7])
    
#add first column
mainl.insert(0, al)

#convert to dataframe
df=pd.DataFrame(mainl).transpose()

#add column names
df.columns=['Kod','Nazwa świadczeniodawcy',"Miasto", "Ulica", "Kod pocztowy", "NIP", "Regon", "Sumaryczna kwota zobowiązań"]

df
```
